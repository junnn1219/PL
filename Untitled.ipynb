```python
import pandas as pd

data = pd.read_csv("hw1111.csv")
data
```

```python
#第一題
#總共幾個基隆人
keelung = data[data['居住地'] == '基隆市']
keelung.counter = len(keelung)
print('共有',keelung.counter,'人')
keelung
```

```python
#第二題
#取交集(是男生且薪水大魚50000的人)
Male_Salary = data[(data['姓別'] == 'M') & (data['薪水'] >50000)]
Male_Salary
```

```python
#第三題
#總共幾個職業是教職人員
job = data[data['職業'] == '教']
job.counter = len(job)
print('共有',job.counter,'人')
job
```

```python
#第四題
#差集(是台北人但是不是11月出生的)
Taipei_Month = data[(data['居住地'] == '台北市') & (data['出生月份'] != 11)]
Taipei_Month
```

```python
#第五題
#誰薪水最高
most_money = data.loc[data['薪水'].idxmax()].name
data.loc[most_money]
```

```python
#第六題
#誰是AB型
AB = data[data['血型'] == 'ＡＢ']
AB.counter = len(AB)
print('共有',AB.counter,'人')
AB
```

```python
#第七題
#誰最年輕
most_youngest = data.loc[data['年紀'].idxmin()].name
data.loc[most_youngest]
```

```python
#第八題
#誰22號出生
bir = data[data['出生日期'] == 22]
bir.counter = len(bir)
print('共有',bir.counter,'人')
bir
```

```python
#第九題
#誰超過30歲
age = data[data['年紀'] > 30 ]
age.counter = len(age)
print('共有',age.counter,'人')
age
```

```python
#第十題
#誰要退休了
age = data[data['年紀'] > 60 ]
age.counter = len(age)
print('共有',age.counter,'人')
age
```
